```python
import numpy as np
import pandas as pd
import cv2
from os import listdir
from PIL import Image
```

- Download UTKFace original at [link](https://www.kaggle.com/datasets/jangedoo/utkface-new)
- Download UTKFace convert to file .csv at [link](https://www.kaggle.com/datasets/lngcthun/utkface-convert-csv)

```python
path = "./UTKFace/"
age = []
gender = []
ethnicity = []
img = []
columns = ["age", "gender", "ethnicity", "image"]
utkface = pd.DataFrame(columns=columns)
for image in listdir(path):
    age.append(image.split('_')[0])
    gender.append(image.split('_')[1])
    ethnicity.append(image.split('_')[2])
    test = cv2.resize(cv2.imread(path+image), dsize = (50,50)).reshape(-1)
    s = ''.join(str(x) + ' ' for x in test).strip()
    img.append(s)
# gg = cv2.resize(cv2.imread('./UTKFace/1_0_2_20161219162518471.jpg.chip.jpg'), dsize = (50,50)).reshape(-1)
# s = ''.join(str(x) + ' ' for x in gg).strip()
# s
```

```python
utkface.age = age
utkface.gender = gender
utkface.ethnicity =ethnicity
utkface.image = img
utkface
```

```python
utkface.to_csv('./utkface.csv')
```

```python
utkface1 = utkface.copy()
utkface1.sort_values(by='age', inplace=True)
utkface1
```
